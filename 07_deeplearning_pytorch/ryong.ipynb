```python
def solution(d):
    n = len(d) - 1
    m = [[0] * n for _ in range(n)]  # 최소 곱셈 횟수를 저장할 배열
    p = [[0] * n for _ in range(n)]  # 최적의 분할점을 저장할 배열
    cost = 0
    for length in range(2, n + 1):
        for i in range(n - length + 1):
            j = i + length - 1
            m[i][j] = float('inf')
            for k in range(i, j):
                cost = m[i][k] + m[k+1][j] + d[i] * d[k+1] * d[j+1]
                if cost < m[i][j]:
                    m[i][j] = cost
                    p[i][j] = k
    return m
solution([[5,3],[3,10]])
```

```python
import sys

def solution(matrix_sizes):
    p = list()
    print(type(matrix_sizes))

    for index1, value1 in enumerate(matrix_sizes):
        p.append(value1[0])
        if index1 == ( len(matrix_sizes) - 1 ):
            p.append(value1[1])
    
    n = len(p) - 1  # 행렬의 개수
    
    # m[i][j]는 행렬 A_i부터 A_j까지 곱하는 최소 곱셈 횟수입니다.
    m = [[0 for _ in range(n)] for _ in range(n)]

    # 길이 1인 행렬 곱셈은 0입니다. (이미 m[i][i]는 0으로 초기화됨)

    # 길이 2부터 n까지 길이(l)를 늘려가며 최적의 곱셈 순서를 찾습니다.
    for l in range(2, n + 1):
        for i in range(n - l + 1):
            j = i + l - 1
            m[i][j] = sys.maxsize  # 최댓값으로 초기화
            
            # i부터 j까지의 행렬을 k를 기준으로 두 그룹으로 나눕니다.
            for k in range(i, j):
                # (A_i...A_k) * (A_{k+1}...A_j)
                # 비용 = (A_i...A_k) 비용 + (A_{k+1}...A_j) 비용 + 두 결과 행렬 곱셈 비용
                cost = m[i][k] + m[k+1][j] + p[i] * p[k+1] * p[j+1]
                if cost < m[i][j]:
                    m[i][j] = cost
    
    # m[0][n-1]에 A_0부터 A_{n-1}까지의 최소 곱셈 횟수가 저장됩니다.
    return m[0][n-1]
solution([[5,3],[3,10],[10,6]])
```
