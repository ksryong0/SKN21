**Langgraph-LLM State Machine**
- 노드-functions, LLM Calls, agents
- 엣지-결과나 상태에 따른 transition(state의 변화)


1. LLM이 스스로 판단해서 다음 단계를 결정하게 만들고 싶은경우(LCEL만으로 부족할 때)

2. 여러 단계를 거치면서 정보(상태)를 계속 유지하고 업데이트 해야하는 경우

3. 작업 결과가 만족스럽지 않으면 다시 이전 단계로 돌아가서 재시도하는 로직이 필요할 때 (순환)

4. 여러 LLM이나 도구들이 상호작용하는 복잡한 워크플로우가 필요한 경우
