```python
# IN, between
# where id in [1, 3, 5]
result = Choice.objects.filter(pk__in=[1, 3, 5])

# where votes between 10 and 30
result = Choice.objects.filter(votes__range=[10, 30])

for c in result:
    print(c, c.votes)

print(result.query)
```

```python
# 조건이 여러개인 경우 - AND
# where id >= 5 and votes < 20
result = Choice.objects.filter(
    pk__gte=5,
    votes__lt=20
)

for c in result:
    print(c, c.votes)

print(result.query)
```

```python
# OR - Q 객체에 조건들을 넣어준다. Q(조건 | 조건 | ..)
from django.db.models import Q

# where votes >= 20 or id <= 10
result = Choice.objects.filter(
    Q(votes__gte=20) | Q(pk__lte=10)
)

result = Choice.objects.filter(
    ~Q(votes__gte=20) | Q(pk__lte=10)
)

for c in result:
    print(c, c.votes)

print(result.query)
```

```python
# 정렬 - QuerySet.order_by("컬럼명", "컬럼명") "-컬럼명": 내림차순(
# result = Choice.objects.all().order_by("votes").reverse()
# order by votes asc id desc
# 컬럼명 앞에 -를 붙이면 내림차순 정렬
result = Choice.objects.all().order_by("votes", "-id")

for c in result:
    print(c, c.votes)

print(result.query)
```

# 집계

- aggregate(집계함수객체(집계할 컬럼명), 집계함수객체(집계할 컬럼명), ...) -> 전체행 기준 집계
- values('groupby 기준 컬럼').annotate(집계함수객체(컬럼명), ..) -> Group by 후 집계

```python
from django.db import models
# select count(id), avg(votes) from table;
result = Choice.objects.aggregate(
    models.Count("id"),
    models.Avg("votes"),
)
# 결과 key를 지정
# select count(id) as cnt, avg(votes) as avg from table;
result = Choice.objects.aggregate(
    cnt = models.Count("id"),
    avg = models.Avg("votes")
)

print(result)
for r in result:
    print(type(r))
```

```python
result['cnt']
```

```python
# Group by 집계
# select min(vote), max(vote)    from choice groupby question
result = Choice.objects.values("question").annotate(
    models.Min('votes'), models.Max('votes')
)

print(type(result))
for r in result:
    print(r)
```
