STS(Security Token Service)
STS 임시 자격 증명 (Access Key, Secret Access Key) : STS API를 호출하여 임시 자격 증명을 발급
다음과 같은 내용이 발급된다.

- Access Key : (AWS에 접근하기 위한) ID와 비슷하다
- Secret Access Key : (AWS에 접근하기 위한) Password와 비슷하다
- Session Token : 임시 보안 자격 증명의 유효성을 확인하는 데 사용되는 토큰, 유효기간이 지나면 만료된다.

STS 임시 자격 증명을 얻는 방법
1. AWS CLI
2. AWS SDK
3. Lambda 함수

STS 유효 시간
(유효기간이 지나면 자동으로 STS가 무효화된다, 다시 갱신 해야됨)

최소(min) : 15 분
최대(max) : 36 시간

주요 STS API
- AssumeRole:특정 역할(Role)을 임시로 가정(Assume)
- AssumRoleWithSAML:회사 조직 내의 SSO 시스템을 사용해 인증된 사용자에게 AWS 리소스 접근을 허용
- AssumeRoleWithWebIdentity:Web Identity Provider(WIP)로부터 인증된 사용자에게 역할을 임시로 가정. WIP는 구글, 페북, 아마존
- GetSessionToken: 사용자가 MFA를 통해 인증후 IAM User의 임시 자격 증명을 요청 가능

자격 증명 기반 정책(Id-Based Policy) : 자격(ID) 기반 정책은 ID(사용자, 사용자 그룹 및 역할)가 수행할 수 있는 작업, 리소스 및 조건을 제어하는 JSON 권한 정책 문서

1. AWS 관리형 정책(AWS Managed Policy): AWS에서 기본으로 제공되는 정책
2. AWS 고객 관리형 정책 AWS Customer Managed Policy) : 말그대로 사용자가 커스텀한거.
3. AWS 인라인 정책 : 하나의 사용자에게 하나의 정책
