IAM(Identity and Access Management)
Identity : Who you are. 인증
Access Management : What you can do. 인가. 권한 관련.

VPC(Virtual Private Cloud)
- 가상 사설 네트워크 환경
- 사용자가 AWS 내부에 가상의 네트워크 환경을 직접 구성하여 사용할 수 있도록 제공하는 서비스

1. Region 선택(데이터센터 그룹의 물리적 위치)
2. IP Range 선택
3. 라우팅 설정
4. Traffic 통제(In/Out)

1. VPC는 Region 단위로 제공되는 서비스임.

2. Region 선택 후 IP 대역을 직접 설정할 수 있음.
Class 방식이 아닌 CIDR(Classless Inter-Domain Routing) 기반. 네트워크의 크기를 / 크기로 유연하게 설정.
CIDR 대역은 VPC 생성 후 변경 불가능
CIDR 대역 결정 기준
- 구축할 서비스 규모
- IP 대역이 많이 필요한 시스템인지
- 향후 확장 가능성
- 타 시스템과 연계 가능성

# 라우트 테이블.
## VPC 생성시 기본으로 하나 제공
## 구성요소
### Destination : 목적지 주소
### Target : 트래픽을 실제로 보내줄 대상. 논리적 리소스의 아이디로 표현(인터넷 게이트웨이는 IGW-xxxx)
### 
### Destination 대역으로 들어온 IP는 Target으로 보내라.
### 대역이 적절하지 않으면 끊겨버림
### 여러 destination이 중복될 때는 /다음이 가장 큰 숫자가 가장 명확

Public 서브넷 : 인터넷 게이트웨이를 통해 외부의 인터넷과 연결되어 있음
Private 서브넷 : 외부 연결 X. 퍼블릭 IP 부여 불가. Public 서브넷이나 VPC 구성요소만 접근 가능. 인터넷 연결 불가. 보안적으로 안전

- 라우트 테이블에서 경로 설정 후 이용 가능.
- 무료.
