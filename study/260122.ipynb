주요 원리 및 장점
- 트래픽 분산: 클라이언트와 서버 그룹 사이에 위치하여 요청을 여러 서버로 고르게 분배합니다.
- 성능 향상: 특정 서버에 트래픽이 몰리는 것을 방지하여 시스템 전체의 응답 속도를 개선합니다.
- 고가용성 및 안정성: 서버 하나에 문제가 생겨도 서비스가 중단되지 않도록 다른 정상 서버로 트래픽을 전달하여 안정성을 높입니다.
- 유연성: 서버를 추가하거나 제거할 때 서비스 중단 없이 유연하게 대응할 수 있습니다. 

서버 확장 방법
- Scale Up:서버 성능을 올림. 하드웨어 업그레이드
- Scale Out:서버 대수를 늘림

주요 로드 밸런싱 알고리즘
- 라운드 로빈 (Round Robin): 요청을 순서대로 서버에 번갈아 분배합니다.
- 가중 라운드 로빈 (Weighted Round Robin): 서버의 처리 능력(가중치)에 따라 요청을 우선적으로 분배합니다. 특정 서버의 성능이 좋을 때 사용.
- 최소 연결 (Least Connections): 현재 가장 적은 연결을 가진 서버로 요청을 보냅니다.
- 최소 응답 (Least Response Time) : 가장 짧은 응답시간을 보이는 서버에 우선적으로 세션을 할당
- IP 해시 (IP Hash): 특정 IP주소나 포트를 갖는 사용자들은 특정 서버로만 세션을 맺도록 하는데, 특정 IP주소나 포트에서 접속량이 특히 많을 때 관리가 편한 방식입니다.

로드 밸런서의 종류
- 소프트웨어 로드 밸런서: NGINX, HAProxy 등 소프트웨어 형태로 구현됩니다.
- 하드웨어 로드 밸런서: F5 Networks의 BIG-IP와 같은 전용 하드웨어 장치로, 높은 성능을 제공하지만 비용이 높습니다.
- 클라우드 로드 밸런서: AWS의 ALB/NLB, Azure Load Balancer 등 클라우드 환경에 최적화된 서비스입니다
